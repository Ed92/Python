```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
#plt.show() if not in Jupyter
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, "r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
# OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title("Smaller plot")
```
